```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

#load the iris dataset
iris=load_iris()
X=iris.data #feature matrix
y=iris.target #true labels for validation

##convert the datframe for easier handling
df=pd.DataFrame(X, columns=iris.feature_names)
```

```python
##preprocess the data
##stardardizing the features
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
#apply k-means
#define the number if clusters
k=3 # we know there are 3 species in the iris dataset
```

```python
#create a kmeans instance
kmeans=KMeans(n_clusters=k, random_state=42)

#fir the model
kmeans.fit(X_scaled)
```

```python
##now get the cluster labels and centroids
labels=kmeans.labels_
centroids=kmeans.cluster_centers_
```

```python
##visulaize the results
##create a dataframe to include the cluster labels
df['Cluster']=labels
```

```python
##now visualize the data by plotting clusters on a scatterplot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df,x='sepal length (cm)', y='sepal width (cm)', hue='Cluster', palette='deep', s=100)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, label='Centroids', color='red')
plt.title('K-Means Clustering on Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
##plt.legend() function is used to display the legend on the plot. in the context of the k means clustering visualization, it will label the centroids and if provided the different clusters.
plt.legend()
plt.show()
```
