```python
from transformers import pipeline
```

```python
##load the sentiment analysis pipeline
classifier=pipeline("sentiment-analysis")
```

```python
##sample texts to analyze
texts=[
    "I love working with NLP medels!"
    "This product is not what I expected."
    "HUgging Face makes NLP so much easier!"
    "This experience was disappointing and frustrating."
]
```

```python
for text in texts:
  result=classifier(text)
  print(f"Text:{text}")
  print(f"Sentiment:{result[0]['label']}, Confidence:{result[0]['score']:.2f}")
  print()
```

```python
##using different NER pipeline
##load the ner pipeline
ner=pipeline("ner", grouped_entities=True)

##example text for the NER
text="HUgging Face was founded in NEW YORK and has headquarters in Paris."

##perform NER
ner_results=ner(text)

##display the results
for entity in ner_results:
  print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Score:{entity['score']:.2f}")
```

```python
##performing question and answering wherein you provide the model a question and a conetext and the model tries to find the answer within the context.
qa_pipeline=pipeline("question-answering")

##now define the context and the question
context="""
Hugging Face is a company that provides open-source tools for machine learning and NLP. They are known for the Transformers library, which are widely used for NLP tasks.
"""

question="What does Hugging Face provide?"

##perform question answering
answer=qa_pipeline(question=question, context=context)

##display the answer
print(f"Question: {question}")
print(f"Answer: {answer['answer']}")
print(f"Confidence:{answer['score']:.2f}")
```

```python
##text summarization: this reduces a long text to its main points. below is an instance of how to use hugging face for summarizing a text.
##example text for summarization.
##load the summarizer pipeline
summarizer=pipeline("summarization")

text="""
Natural Language Processing(NLP) is a subfield of linguistics,
computer science, and artificial intelligence concerned with the interactions between
computers and human language, in particular how to program computers to process
and analyze large amounts of natural language data. The goal is a computer capable
of "understanding" the contents of documents, including the contextual nuances of the language within them.
"""

summary=summarizer(text, max_length=50, min_length=25, do_sample=False)
print("Summary:", summary[0]['summary_text'])
```

```python
##combining all these task in a single function
def perform_nlp_task(text, task="sentiment-analysis"):
    # Initialize a specific NLP pipeline based on the task
    nlp_task = pipeline(task)

    if task == "sentiment-analysis":
        result = nlp_task(text)
    elif task == "ner":
        result = nlp_task(text, grouped_entities=True)
    elif task == "question-answering":
        question, context = text  # Input is a tuple (question, context)
        result = nlp_task(question=question, context=context)
    elif task == "summarization":
        result = nlp_task(text, max_length=50, min_length=25, do_sample=False)
    else:
        raise ValueError("Unsupported task")

    return result

# Example usage
print("Sentiment Analysis:", perform_nlp_task("I love NLP!"))
print("NER:", perform_nlp_task("Hugging Face is based in New York and Paris.", task="ner"))
print("QA:", perform_nlp_task(("What does Hugging Face provide?", "Hugging Face provides NLP tools."), task="question-answering"))
print("Summarization:", perform_nlp_task("NLP is a field of AI that focuses on understanding human language...", task="summarization"))
```
